# Minmax scaler 0-100

```python
from sklearn.preprocessing import MinMaxScaler
# Tạo bản sao để tránh ảnh hưởng tới df gốc
df2 = df.copy()

# Bước 1: Lọc các cột cần chuẩn hóa (tất cả các cột trừ 'Country', 'year' và 'SDG Index Score')
cols_to_scale = [col for col in df2.columns if col not in ['Country', 'year', 'SDG Index Score']]

# Bước 2: Chuẩn hóa các cột đã lọc
scaler = MinMaxScaler(feature_range=(0, 100))
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])

# In kết quả
print(df2.head())
```

# Model dự đoán targets

```python
df3=df2.copy()
```

## Tạo lag

Note: Biến lag phản ánh sự thay đổi giá trị của các biến theo thời gian và quốc gia. Vì time series không dùng được, ML không giỏi với dữ liệu dạng time series nên biến lag là hướng giải quyết để tạo thêm ảnh hưởng về thời gian vào mô hình ML
(này tớ viết theo kiểu tớ hiểu, ngôn từ có thể có sai sót, mn check lại nha)

```python
# 1. Tạo biến lag cho tất cả các cột trừ 'Country', 'year', 'SDG Index Score'
exclude_cols = ['Country', 'year', 'SDG Index Score']
lag_cols = [col for col in df3.columns if col not in exclude_cols]

# Tạo các cột lag
for col in lag_cols:
    df3[f'{col}_lag'] = df3.groupby('Country')[col].shift(1)
```

```python
# 2. Loại bỏ các dòng có giá trị NaN ở bất kỳ cột lag nào
lagged_cols = [f'{col}_lag' for col in lag_cols]
df3 = df3.dropna(subset=lagged_cols).reset_index(drop=True)
```

```python
# 3. Mã hóa cột Country thành số
df3['Country_ID'] = df3['Country'].astype('category').cat.codes

df3
```

## Train test

Note: Chia tập train test theo năm để bảo đảm tính chất thời gian

```python
# Chia dữ liệu train/test theo năm
train_data = df3[df3['year'] <= 2020].copy()
test_data = df3[(df3['year'] >= 2021) & (df3['year'] <= 2024)].copy()

print(f"Số dòng train: {len(train_data)}, test: {len(test_data)}")
```

## Model

Dùng 3 mô hình Linear Regression đại diện cho các mô hình tuyến tính, Random forest - Ensemble Bagging, XGBoost - Ensemble Boosting

Chưa GridSearchCV tìm tham số + Cross Validation

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error

# Lặp qua từng Target để huấn luyện mô hình
models = {}  # Lưu các mô hình đã huấn luyện

X_train = train_data.drop(['Country', 'year','SDG Index Score'], axis=1).copy()
y_train = train_data['SDG Index Score'].copy()

model_lr = LinearRegression()
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_xgb = xgb.XGBRegressor(n_estimators=100, random_state=42)
    
# Huấn luyện mô hình Linear Regression
model_lr.fit(X_train, y_train)
models[f'{X_train}_lr'] = model_lr
```

```python
from sklearn.model_selection import GridSearchCV

# Lưới siêu tham số cho Random Forest
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# GridSearchCV cho Random Forest
grid_rf = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid_rf,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_rf.fit(X_train, y_train)
models['rf_best'] = grid_rf.best_estimator_
print("Best parameters for RF:", grid_rf.best_params_)
```

```python
# Lưới siêu tham số cho XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.2]
}

# GridSearchCV cho XGBoost
grid_xgb = GridSearchCV(
    estimator=xgb.XGBRegressor(random_state=42, verbosity=0),
    param_grid=param_grid_xgb,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_xgb.fit(X_train, y_train)
models['xgb_best'] = grid_xgb.best_estimator_
print("Best parameters for XGB:", grid_xgb.best_params_)
```

```python
# Sau khi chạy GridSearchCV
model_rf = grid_rf.best_estimator_
model_xgb = grid_xgb.best_estimator_
```

## Validate test

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Chuẩn bị X_test và y_test
X_test = test_data.drop(['Country', 'year','SDG Index Score'], axis=1).copy()
y_test = test_data['SDG Index Score'].copy()

# Dự đoán trên tập test
y_pred_lr = model_lr.predict(X_test)
y_pred_rf = model_rf.predict(X_test)
y_pred_xgb = model_xgb.predict(X_test)

# Đánh giá mô hình trên tập test
model_preds = {
    'Linear Regression': y_pred_lr,
    'Random Forest': y_pred_rf,
    'XGBoost': y_pred_xgb
}

for model_name, y_pred in model_preds.items():
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f'--- {model_name} ---')
    print(f'Mean Squared Error (MSE): {mse:.5f}')
    print(f'Mean Absolute Error (MAE): {mae:.5f}')
    print(f'R^2 Score: {r2:.5f}\n')
```

## Forecasting

### Forecasting features of VN

Phần này dự đoán các biến, kể cả biến lag và cho ra 1 df bao gồm các giá trị dự đoán của các cột từ năm 2025-2030

```python
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

# Danh sách các cột target 
target_columns = df3.drop(['Country', 'year', 'SDG Index Score'], axis=1)

country_of_interest = 'Vietnam'
history_end_year = 2024
forecast_start_year = 2025
forecast_end_year = 2030
```

```python
# Dự báo các Biến Ngoại sinh cho Việt Nam (2025-2030) ---

# Lọc dữ liệu lịch sử cho Việt Nam
vietnam_history_df = df3[df3['Country'] == 'Vietnam'].copy()

# DataFrame để lưu kết quả dự báo targets
forecasted_targets_vn = pd.DataFrame(index=range(forecast_start_year, forecast_end_year+1))

for target_col in target_columns:
    print(f"  - Dự báo Target: {target_col}")
    # Lấy dữ liệu target lịch sử cho Việt Nam
    target_history = vietnam_history_df[['year', target_col]].dropna()

    if not target_history.empty:
        # Sử dụng statsmodels OLS để fit Linear Regression trên thời gian (có thêm hằng số)
        X_year = sm.add_constant(target_history['year'])
        y_target = target_history[target_col]

        try:
            lr_target_model = sm.OLS(y_target, X_year).fit()

            # Tạo dữ liệu năm cho giai đoạn dự báo
            future_years_df = pd.DataFrame({'year': range(forecast_start_year, forecast_end_year+1)})
            X_future_year = sm.add_constant(future_years_df['year'])

            # Dự báo giá trị target cho tương lai
            forecasted_targets_vn[target_col] = lr_target_model.predict(X_future_year).values

        except Exception as e:
             print(f"    Lỗi khi fit Linear Regression cho {target_col}: {e}. Cột này có thể không được dự báo.")
             forecasted_targets_vn[target_col] = np.nan 

    else:
        print(f"    Cảnh báo: Không có dữ liệu lịch sử cho {target_col} của Việt Nam. Cột này không được dự báo.")
        forecasted_targets_vn[target_col] = np.nan 

# Xử lý các cột target không dự báo được nếu có
forecasted_targets_vn = forecasted_targets_vn.dropna(axis=1, how='all')
target_columns_forecasted = forecasted_targets_vn.columns.tolist()

print("\n--- Kết quả Dự báo Targets (2025-2030) ---")
print(forecasted_targets_vn)
print("-" * 30)
```

# Phân tích với SHAP

Phần phân tích với SHAP chủ yếu làm rõ hơn các target ảnh hưởng tới SDG Index Score, ảnh hưởng cụ thể thế nào,.. Phần này nằm trước phần chạy model

## Tính toán SHAP

```python
# Chia dữ liệu train/test theo năm
train_data_shap = df2[df2['year'] <= 2020].copy()
test_data_shap = df2[(df2['year'] >= 2021) & (df2['year'] <= 2024)].copy()
```

```python
X_train_shap = train_data_shap.drop(['Country', 'year','SDG Index Score'], axis=1).copy()
y_train_shap = train_data_shap['SDG Index Score'].copy()
```

```python
import shap
import xgboost as xgb

shap.initjs()

# Vì 'SDG Index Score' là giá trị liên tục nên ta dùng Regressor
model = xgb.XGBRegressor(objective='reg:squarederror', 
                         n_estimators=100,
                         learning_rate=0.1,
                         random_state=42)

model.fit(X_train_shap, y_train_shap)
```

```python
X_test_shap = test_data_shap.drop(['Country', 'year','SDG Index Score'], axis=1).copy()
y_test_shap = test_data_shap['SDG Index Score'].copy()
```

```python
# Tạo SHAP explainer cho mô hình cây (XGBoost là mô hình cây)
# shap.TreeExplainer được tối ưu cho các mô hình cây
explainer = shap.TreeExplainer(model)

# Tính toán giá trị SHAP cho tập kiểm tra
# shap_values sẽ là một mảng numpy với kích thước (số lượng mẫu test, số lượng đặc trưng)
shap_values = explainer.shap_values(X_test_shap)

print("\nĐã tính toán giá trị SHAP cho tập kiểm tra.")
```

## Summary Plot

```python
# 1. Summary Plot: Tổng quan về ảnh hưởng của các đặc trưng
# Biểu đồ này hiển thị mức độ quan trọng tổng thể của mỗi đặc trưng và hướng ảnh hưởng của nó.
# Màu sắc: đỏ (giá trị đặc trưng cao), xanh dương (giá trị đặc trưng thấp).
# Trục x: giá trị SHAP (ảnh hưởng đến dự đoán).
print("\nHiển thị Summary Plot:")
shap.summary_plot(shap_values, X_test_shap)
```

## Force Plot

Force Plot giải thích SDG Score cho việt nam năm 2024, tức là, điểm số SDG Score của VN bị ảnh hưởng bao nhiêu % bởi các chỉ số nào

```python
# 2. Force Plot (cho một dự đoán cá nhân): Giải thích một dự đoán cụ thể
sample_idx = len(X_test_shap) - 1
print(f"\nHiển thị Force Plot cho mẫu dữ liệu thứ {sample_idx}:")
shap.force_plot(explainer.expected_value, shap_values[sample_idx,:], X_test_shap.iloc[sample_idx,:])
```

## Dependence Plot

 Dependence Plot dùng để khám phá cách MỘT ĐẶC TRƯNG CỤ THỂ ảnh hưởng đến dự đoán của mô hình TRÊN TOÀN BỘ TẬP DỮ LIỆU, và liệu mối quan hệ đó có bị thay đổi bởi một đặc trưng khác hay không.

```python
# 3. Dependence Plot: Khám phá mối quan hệ giữa một đặc trưng và ảnh hưởng của nó
feature_to_plot = pd.Series(np.mean(np.abs(shap_values), axis=0), index=X_test_shap.columns).sort_values(ascending=False).head(10).index.tolist()
for col in feature_to_plot:
    print(f"\nHiển thị Dependence Plot cho đặc trưng '{col}':")
    shap.dependence_plot(col, shap_values, X_test_shap)
```

## Waterfall Plot

Cũng na ná Force Plot nhma trực quan theo kiểu Waterfall Plot - dữ liệu Việt nam năm 2024

```python
# 4. Waterfall Plot (cho một dự đoán cá nhân): Trình bày chi tiết đóng góp của từng đặc trưng
print(f"\nHiển thị Waterfall Plot cho mẫu dữ liệu thứ {sample_idx}:")
shap_object = shap.Explanation(values=shap_values[sample_idx],
                               base_values=explainer.expected_value,
                               data=X_test.iloc[sample_idx],
                               feature_names=X_test_shap.columns.tolist())

shap.waterfall_plot(shap_object)
```
